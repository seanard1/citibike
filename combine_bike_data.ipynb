```python
import pandas as pd
from pathlib import Path
from datetime import datetime
```

# 2019 Data


```python
# Starting building for a for loop to process, but because of data reporting change in 2020 to 2021, I realized just five 
# separate cleans would be faster for the coding. That is why there is pointless string manipulation for file path.

# Really glad I did this, because the July data I started with didn't look correct for 2023, so all I did to switch
# to June was change the month once. 

prefix = 'citibike-tripdata/20'
year = 190
month = 6
suffix = '-citibike-tripdata.csv'


#Read in the 2019 CSV

data_path = prefix + str(year) + str(month) + suffix

data_path

bike_2019_df = pd.read_csv(
    Path(data_path)
)

bike_2019_df
    
```

```python
# Reduce 2.1 million trips down to those only equal to or greater than 15 minutes by subscribers. Result is ~500,000 trips

bike_2019_df = bike_2019_df.loc[(bike_2019_df['tripduration'] >= 900) & (bike_2019_df['usertype'] == 'Subscriber')  & (bike_2019_df['tripduration'] <= 86400)]
bike_2019_df
```

```python
# Save to a CSV for Tableau

bike_2019_df.to_csv(
        Path('data/2019-data.csv'), index=False
    )
```

# 2020 Data


```python
# Same ultimately pointless string concat for Path. Keeping in case I do for loop later. 

year = 200
data_path = prefix + str(year) + str(month) + suffix

data_path

bike_2020_df = pd.read_csv(
    Path(data_path)
)

bike_2020_df
```

```python
# Filters this time take it from 2.1 million to just shy of 400,000 trips.

bike_2020_df = bike_2020_df.loc[(bike_2020_df['tripduration'] >= 900) & (bike_2020_df['usertype'] == 'Subscriber')  & (bike_2020_df['tripduration'] <= 86400)]
bike_2020_df
```

```python
# Save to CSV

bike_2020_df.to_csv(
        Path('data/2020-data.csv'), index=False
    )
```

# 2021 Data

```python
# Same import idea with string manipulation.

year = 210
data_path = prefix + str(year) + str(month) + suffix

data_path

bike_2021_df = pd.read_csv(
    Path(data_path)
)

bike_2021_df
```

```python
# This time, due to data reporting change, we need to calculate the trip duration ourselves before applying filter.


# These lines change the start and end times to datetime format
bike_2021_df['started_at'] = pd.to_datetime(bike_2021_df['started_at'], format="%Y-%m-%d %H:%M:%S.%f")
bike_2021_df['ended_at'] = pd.to_datetime(bike_2021_df['ended_at'], format="%Y-%m-%d %H:%M:%S.%f")

# This adds another column with the calculated trip duration. 
bike_2021_df['tripduration'] = (bike_2021_df['ended_at'] - bike_2021_df['started_at']).dt.total_seconds().astype('int64')

# And now we apply the same filter as before. Note that 'Subscriber' became 'member' in a renamed column.
bike_2021_df = bike_2021_df.loc[(bike_2021_df['tripduration'] >= 900) & (bike_2021_df['member_casual'] == 'member') & (bike_2021_df['tripduration'] <= 86400)]
bike_2021_df
```

```python
# Save to CSV

bike_2021_df.to_csv(
        Path('data/2021-data.csv'), index=False
    )
```

# Attempt to merge all 

```python
# Organize all the dataframes to be the same // Code is same for 2019 and 2020, then altered for the group of 2021, 2022 and 2023

bike_2019_df['starttime'] = pd.to_datetime(bike_2019_df['starttime'], format="%Y-%m-%d %H:%M:%S.%f")
bike_2019_df['stoptime'] = pd.to_datetime(bike_2019_df['stoptime'], format="%Y-%m-%d %H:%M:%S.%f")

bike_2019_df.rename(columns={'starttime': 'started_at', 'stoptime': 'ended_at', 'usertype': 'member_casual'}, inplace=True)
bike_2019_df.rename(columns={'start station id': 'start_station_id', 'start station name': 'start_station_name', 'start station latitude': 'start_lat', 'start station longitude': 'start_lng'}, inplace=True)
bike_2019_df.rename(columns={'end station id': 'end_station_id', 'end station name': 'end_station_name', 'end station latitude': 'end_lat', 'end station longitude': 'end_lng'}, inplace=True)
bike_2019_df.drop(columns=['bikeid', 'birth year', 'gender', 'member_casual', 'start_station_id', 'end_station_id'], inplace=True)
bike_2019_df.rename(columns={'tripduration': 'trip_duration'}, inplace=True)

bike_2019_df.info()
```

```python
# Organize all the dataframes to be the same

bike_2020_df['starttime'] = pd.to_datetime(bike_2020_df['starttime'], format="%Y-%m-%d %H:%M:%S.%f")
bike_2020_df['stoptime'] = pd.to_datetime(bike_2020_df['stoptime'], format="%Y-%m-%d %H:%M:%S.%f")

bike_2020_df.rename(columns={'starttime': 'started_at', 'stoptime': 'ended_at', 'usertype': 'member_casual'}, inplace=True)
bike_2020_df.rename(columns={'start station id': 'start_station_id', 'start station name': 'start_station_name', 'start station latitude': 'start_lat', 'start station longitude': 'start_lng'}, inplace=True)
bike_2020_df.rename(columns={'end station id': 'end_station_id', 'end station name': 'end_station_name', 'end station latitude': 'end_lat', 'end station longitude': 'end_lng'}, inplace=True)
bike_2020_df.drop(columns=['bikeid', 'birth year', 'gender', 'member_casual', 'start_station_id', 'end_station_id'], inplace=True)
bike_2020_df.rename(columns={'tripduration': 'trip_duration'}, inplace=True)

bike_2020_df.info()
```

```python
# Organize all the dataframes to be the same

bike_2021_df.rename(columns={'tripduration': 'trip_duration'}, inplace=True)
bike_2021_df.drop(columns=['ride_id', 'rideable_type'], inplace=True)
bike_2021_df = bike_2021_df[['trip_duration', 'started_at', 'ended_at', 'start_station_name', 'start_lat', 'start_lng', 'end_station_name', 'end_lat', 'end_lng']]
bike_2021_df.info()
```

```python
# Organize all the dataframes to be the same

bike_2022_df.rename(columns={'tripduration': 'trip_duration'}, inplace=True)
bike_2022_df.drop(columns=['ride_id', 'rideable_type'], inplace=True)
bike_2022_df = bike_2022_df[['trip_duration', 'started_at', 'ended_at', 'start_station_name', 'start_lat', 'start_lng', 'end_station_name', 'end_lat', 'end_lng']]
bike_2022_df.info()
```

```python
# Organize all the dataframes to be the same

bike_2023_df.rename(columns={'tripduration': 'trip_duration'}, inplace=True)
bike_2023_df.drop(columns=['ride_id', 'rideable_type'], inplace=True)
bike_2023_df = bike_2023_df[['trip_duration', 'started_at', 'ended_at', 'start_station_name', 'start_lat', 'start_lng', 'end_station_name', 'end_lat', 'end_lng']]
bike_2023_df.info()
```

```python
# Drop Nulls

bike_2019_df.dropna(inplace=True)
bike_2020_df.dropna(inplace=True)
bike_2021_df.dropna(inplace=True)
bike_2022_df.dropna(inplace=True)
bike_2023_df.dropna(inplace=True)
```

```python
# Grouby by current (2023) station names and find mean of lat/long

lats = bike_2023_df.groupby('start_station_name')['start_lat'].mean().to_dict()
longs = bike_2023_df.groupby('start_station_name')['start_lng'].mean().to_dict()

# print(lats)

#Apply those new lat longs across all datasets

bike_2019_df['start_lat'] = bike_2019_df['start_station_name'].map(lats)
bike_2019_df['start_lng'] = bike_2019_df['start_station_name'].map(longs)

bike_2020_df['start_lat'] = bike_2020_df['start_station_name'].map(lats)
bike_2020_df['start_lng'] = bike_2020_df['start_station_name'].map(longs)

bike_2021_df['start_lat'] = bike_2021_df['start_station_name'].map(lats)
bike_2021_df['start_lng'] = bike_2021_df['start_station_name'].map(longs)

bike_2022_df['start_lat'] = bike_2022_df['start_station_name'].map(lats)
bike_2022_df['start_lng'] = bike_2022_df['start_station_name'].map(longs)

bike_2023_df['start_lat'] = bike_2023_df['start_station_name'].map(lats)
bike_2023_df['start_lng'] = bike_2023_df['start_station_name'].map(longs)
```

```python
# Make one giant dataset for all

bike_all_df = pd.concat([bike_2019_df, bike_2020_df, bike_2021_df, bike_2022_df, bike_2023_df])
```

```python
# DropNAs again, as some no longer existing stations remained.

bike_all_df.dropna(inplace=True)
```

```python
# Save to CSV

bike_all_df.to_csv(
        Path('data/all-data.csv'), index=False
    )
```
